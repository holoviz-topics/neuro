```python
import pandas as pd
import numpy as np
import panel as pn; pn.extension('tabulator')

vals = np.arange(100)
df = pd.DataFrame({f'column{l}':vals for l in np.arange(14)})

widget = pn.WidgetBox(
    pn.widgets.Tabulator(df,
                         layout='fit_columns', # this does not shrink less than length of column str
                         # [fit_data, fit_data_fill, fit_data_stretch, fit_data_table, fit_columns]'
                         pagination='local',
                         page_size=10,
                         # widths={f'{c}': f'{100/(df.shape[1] + 1)}%' for c in df.reset_index().columns},
                         sizing_mode='stretch_width'
                        ),
)

template = pn.template.BootstrapTemplate(
    sidebar=[widget],
)
template.servable();
```
